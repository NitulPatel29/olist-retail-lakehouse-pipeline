```python
dbutils.widgets.text("sf_user", "")
dbutils.widgets.text("sf_password", "")
dbutils.widgets.text("sf_account", "vzzrrpv-il44046")
```

```python
sf_user = dbutils.widgets.get("sf_user")
sf_password = dbutils.widgets.get("sf_password")
sf_account = dbutils.widgets.get("sf_account")
```

```python
sfOptions = {
    "sfaccount": sf_account,
    "host": f"{sf_account}.snowflakecomputing.com",
    "sfUser": sf_user,
    "sfPassword": sf_password,
    "sfDatabase": "RETAIL_DB",
    "sfSchema": "GOLD",
    "sfWarehouse": "COMPUTE_WH"
}

gold_path = "/Volumes/workspace/retail_raw/gold/"

daily_revenue = spark.read.format("delta").load(gold_path + "daily_revenue")
revenue_by_state = spark.read.format("delta").load(gold_path + "revenue_by_state")
top_categories = spark.read.format("delta").load(gold_path + "top_categories")

print("Gold tables loaded, ready to push to Snowflake")
```

```python
daily_revenue.write \
    .format("snowflake") \
    .options(**sfOptions) \
    .option("dbtable", "DAILY_REVENUE") \
    .mode("overwrite") \
    .save()

print("✅ daily_revenue pushed to Snowflake")
```

```python
revenue_by_state.write \
    .format("snowflake") \
    .options(**sfOptions) \
    .option("dbtable", "REVENUE_BY_STATE") \
    .mode("overwrite") \
    .save()

print("✅ revenue_by_state pushed to Snowflake")

top_categories.write \
    .format("snowflake") \
    .options(**sfOptions) \
    .option("dbtable", "TOP_CATEGORIES") \
    .mode("overwrite") \
    .save()

print("✅ top_categories pushed to Snowflake")
```
